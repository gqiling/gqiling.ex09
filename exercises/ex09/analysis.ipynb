**Step 3 — Define a custom helper function to filter and convert Likert columns.**

Survey Likert responses are optional, so some rows have empty strings for our key numeric columns. Our helper function `filter_numeric_rows` keeps only rows where all specified numeric columns are non-empty, and converts those columns from `str` to `int`. Non-numeric columns (like `primary_major`) remain as strings. This is our required custom helper function for the analysis.

```python
def filter_numeric_rows(
    table: dict[str, list[str]],
    numeric_cols: list[str]
) -> dict[str, list[str | int]]:
    """Return a new table keeping only rows where ALL numeric_cols are non-empty.
    
    Numeric columns are converted from str to int in the output.
    Non-numeric columns remain as strings.
    
    Args:
        table:        A column-oriented data table (all values str).
        numeric_cols: Column names that must be non-empty and will be converted to int.
    
    Returns:
        A new column-oriented table with filtered and type-converted data.
    """
    num_rows: int = len(next(iter(table.values())))
    keep_indices: list[int] = []

    for i in range(num_rows):
        if all(table[col][i] != "" for col in numeric_cols):
            keep_indices.append(i)

    result: dict[str, list[str | int]] = {}
    for col in table:
        if col in numeric_cols:
            result[col] = [int(table[col][i]) for i in keep_indices]
        else:
            result[col] = [table[col][i] for i in keep_indices]

    return result


# Apply the filter: keep only rows with valid Likert scores in all four key columns
numeric_likert_cols: list[str] = ["interested_connections", "interesting", "would_recommend", "valuable"]
clean_data: dict[str, list[str | int]] = filter_numeric_rows(analysis_data, numeric_likert_cols)

print(f"Rows remaining after removing empty Likert responses: {len(clean_data['interested_connections'])}")
```

**Chart 4 — Line plot: Mean `interesting`, `would_recommend`, and `valuable` scores by `interested_connections` level.**

If all three climb together, the case for cross-disciplinary course design is much stronger.

We first build a "long-format" version of the data manually (without pandas) so `seaborn` can distinguish the three metrics using `hue`.

```python
# Build a grouped-average table manually (without pandas)
# We need: for each interested_connections score (1-7),
# compute the mean of interesting, would_recommend, and valuable.

def mean_by_group(
    group_col: list[int],
    value_col: list[int],
    group_values: list[int]
) -> list[float]:
    """Compute the mean of value_col for each group level in group_values."""
    means: list[float] = []
    for g in group_values:
        vals: list[int] = [value_col[i] for i in range(len(group_col)) if group_col[i] == g]
        if len(vals) > 0:
            means.append(round(sum(vals) / len(vals), 3))
        else:
            means.append(0.0)
    return means


# Groups: scores 1 through 7
conn_scores: list[int] = [1, 2, 3, 4, 5, 6, 7]

mean_interesting:     list[float] = mean_by_group(clean_data["interested_connections"], clean_data["interesting"],     conn_scores)
mean_recommend:       list[float] = mean_by_group(clean_data["interested_connections"], clean_data["would_recommend"], conn_scores)
mean_valuable:        list[float] = mean_by_group(clean_data["interested_connections"], clean_data["valuable"],        conn_scores)

# Build a long-format table manually for seaborn
long_x:       list[int]   = conn_scores * 3
long_y:       list[float] = mean_interesting + mean_recommend + mean_valuable
long_metric:  list[str]   = (["interesting"] * 7 + ["would_recommend"] * 7 + ["valuable"] * 7)

long_data: dict[str, list] = {
    "interested_connections": long_x,
    "Mean Score": long_y,
    "Metric": long_metric,
}

# Chart 4: line plot
chart4: sns.FacetGrid = sns.relplot(
    data=long_data,
    kind="line",
    x="interested_connections",
    y="Mean Score",
    hue="Metric",
    marker="o",
    palette="Set2",
    height=5,
    aspect=1.4
)
chart4.set(
    title="Mean Course Sentiment Scores\nby Level of Interest in CS–Other Field Connections",
    xlim=(0.5, 7.5),
    ylim=(1, 7),
    xticks=[1, 2, 3, 4, 5, 6, 7]
)
chart4.set_axis_labels(
    "Interest in CS Connections to Other Fields (1–7)",
    "Mean Likert Score (1–7)"
)
chart4.legend.set_title("Metric")
plt.savefig("chart4_mean_scores_by_connections.png", dpi=150, bbox_inches="tight")
plt.show()
```

This supports designing or re-framing existing programming exercises around students' own fields of study.

#### Recommendation

The course should pilot at least two or three programming exercises per semester where students can choose a dataset or problem domain relevant to their own primary major. Given that non-CS majors are already among the most cross-disciplinarily curious students (Chart 1), they are the group most likely to benefit from this change.

#### Potential Costs, Trade-offs, and Negatively Impacted Stakeholders

- **Instructional staff workload:** Creating and maintaining multiple domain-specific exercise variants requires significant upfront design effort. Grading heterogeneous submissions may also be more complex.
- **Risk of diluting core CS content:** If domain exercises lean too far into the application domain, students may complete them without deeply understanding the underlying programming concepts — the actual learning objective of the course.
- **Equity of options:** Students who are "Undecided" in their major or whose field has few obvious CS applications may feel their option is less engaging than those of peers in biology or business.
- **Correlation ≠ causation:** Students who are already curious about connections may simply tend to find everything more interesting — the intervention itself may not produce the gains we observe in the correlational data.
